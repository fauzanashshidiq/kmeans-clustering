### Import Library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

### load data

```python
baca = pd.read_csv("go_track_tracks.csv")
baca.head()
```

#### informasi tentang dataset

```python
baca.info()
```

#### drop kolom "linha"

```python
baca = baca.drop(["linha"], axis=1)
baca.head()
```

#### penentuan variabel yang diklusterkan

```python
#Menentukan variabel yang akan di klusterkan
baca_x = baca.iloc[:, 1:3]
baca_x.head()

#Memvisualkan persebaran data
plt.scatter(baca.distance, baca.speed, s =10, c = "red", marker = "o", alpha = 0.5)
plt.show()
```

#### mengubah variabel data frame menjadi array

```python
x_array = np.array(baca_x)
print(x_array)
```

### normalisasi data menggunakan metode Min-Max Scaling

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_array)
x_scaled
```

### Membuat Model Clustering

```python
from sklearn.cluster import KMeans

# Membuat model KMeans
kmeans = KMeans(n_clusters=3, random_state=42)

# Training model
kmeans.fit(baca)

# Menampilkan label cluster
print(kmeans.labels_)

# Menambahkan hasil cluster ke dataframe
baca["kluster"] = kmeans.labels_

# Menampilkan data
print(baca.head())
```

### Membuat Report Hasil Clustering

```python
from sklearn.metrics import silhouette_score

# Silhoutte Score
sil_score = silhouette_score(baca, kmeans.labels_)
print(f"Silhouette Score      : {sil_score:.4f}")
```

### Visualisasi dari kluster

```python
import matplotlib.pyplot as plt

# Plot hasil clustering
plt.figure(figsize=(8,6))

plt.scatter(
    baca['id_android'],
    baca['speed'],
    c=baca['kluster'],
    cmap='viridis'
)

# Titik centroid
plt.scatter(
    kmeans.cluster_centers_[:, 1],   # Frequency
    kmeans.cluster_centers_[:, 2],   # Monetary
    s=200,
    c='red',
    marker='X',
    label='Centroid'
)

plt.xlabel('Frequency')
plt.ylabel('Monetary')
plt.title('K-Means Clustering')
plt.legend()

plt.show()
```
